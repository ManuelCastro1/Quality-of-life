```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go


datos = pd.read_csv("ejemplolibertadeconomica.csv")
pais = datos["País"].values
calidad_de_vida = datos["Índice de Calidad de Vida"].str.replace(",",".").astype(float)
costo_de_vida = datos["Índice de Costo de Vida"].str.replace(",",".").astype(float)

correlacion = np.corrcoef(calidad_de_vida,costo_de_vida)
correlacion = correlacion[0,1]

fig = go.Figure()
fig.add_trace(
    go.Scatter(x=calidad_de_vida, y = costo_de_vida, mode="markers",text=pais)
    )

factor = np.min(costo_de_vida)-np.min(calidad_de_vida)

fig.add_trace(
    go.Scatter(x=calidad_de_vida, y = (calidad_de_vida*correlacion+factor))
)
fig.update_traces(textposition='top center')

print(np.where(pais=="Nigeria"))

fig.add_annotation(x=calidad_de_vida[67], y=costo_de_vida[67],   text="Colombia", showarrow=True, arrowhead=1)
fig.add_annotation(x=calidad_de_vida[1], y=costo_de_vida[1],   text="Dinamarca", showarrow=True, arrowhead=1)
fig.add_annotation(x=calidad_de_vida[82], y=costo_de_vida[82],   text="Nigeria", showarrow=True, arrowhead=1)

fig.update_xaxes(title="Índice de Calidad de Vida")
fig.update_yaxes(title="Índice de Costo de Vida")
fig.update_layout(showlegend=False)
fig.update_traces(selector=dict(type='scatter'), showlegend=False)

fig.show()
```

```python
datoslib = pd.read_csv("ejemplolibertadeconomica2.csv")
 
paises = datoslib.values
paises = paises[:,0]
listapaises = []
a = 0
while a < len(paises):
  pais = paises[a]
  if isinstance(pais,str):
    pais = pais.split("\xa0", 1)[0]
    listapaises.append(pais)
  
  
  a = a +1
 
datos = pd.read_csv("ejemplolibertadeconomica.csv")
pais = datos["País"].values
calidad_de_vida = datos["Índice de Calidad de Vida"].str.replace(",",".").astype(float)
 
listalibertad = []
paisesdef = []
calidadvida = []
i = 0
while i < len(listapaises):
  a = 0
  while a < len(pais):
    if listapaises[i] ==  pais[a]:
      listalibertad.append(datoslib.values[i,1])
      paisesdef.append(listapaises[i])
      calidadvida.append(calidad_de_vida[a].astype(float))
      
 
      a = len(pais)
    a = a +1
  i = i +1
 
 
 
listalibertad = np.asarray(listalibertad)
calidadvida = np.asarray(calidadvida)
correlacion = np.corrcoef(calidadvida,listalibertad)
correlacion = correlacion[0,1]
 
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=listalibertad, y = calidadvida, mode="markers",text=paisesdef)
    )
 
 
factor = (np.max(calidadvida)-np.min(calidadvida))/2
fig.add_trace(
    go.Scatter(x=listalibertad, y = (listalibertad*correlacion+factor))
)
 
fig.update_xaxes(title="Índice de libertad económica")
fig.update_yaxes(title="Índice de Calidad de Vida")
fig.update_layout(showlegend=False)
fig.update_traces(selector=dict(type='scatter'), showlegend=False)
 
fig.show()
 
```

```python
import math

idhdat = pd.read_csv("Human Development Index (HDI).csv")
freedom = pd.read_csv("data.csv")
names = freedom.values[:,0]
countries = []
freedomindex = []
idh = []

a = 0
while a < len(names):
  name = str(names[a])
  #name = name.split(" ", 1)[0]
  j = 0
  while j < len(idhdat):
    found = str(idhdat.values[j,0])
    found = found[1:]
    

    if found == name and math.isnan(float(freedom.values[a,2])) == False:
      countries.append(name)
      freedomindex.append(float(freedom.values[a,2]))
      idh.append(float(idhdat.values[j,1]))

      
      j = len(idhdat)


      

    j = j +1

  a = a + 1


idh = (100*(np.asarray(idh).astype(float)))
freedomindex = (np.asarray(freedomindex).astype(float))

fig = go.Figure()
fig.add_trace(
    go.Scatter(x = freedomindex, y = idh, text=countries, mode="markers")
)


correlacion = np.corrcoef(freedomindex,idh)
correlacion = correlacion[0,1]
print(correlacion)



factor = (np.max(idh)-np.min(idh))/2

fig.add_trace(
    go.Scatter(x=freedomindex, y = (freedomindex*correlacion+factor), mode="lines")
)



fig.add_annotation(x=69.7, y=76.3,   text="Colombia", showarrow=True,  font=dict(family="Arial", size=16, color="#000000" ),   align="center",      arrowhead=2,        arrowsize=1,        arrowwidth=2,        arrowcolor="#000000",        ax=-10,        ay=-40,        bordercolor="#FFFFFF",        borderwidth=2,        borderpad=4,        bgcolor="#FFEA35",        opacity=0.99)
fig.add_annotation(x=75.1, y=93.6,   text="Dinamarca", showarrow=True,font=dict(family="Arial", size=16, color="#000000" ),   align="center",      arrowhead=2,        arrowsize=1,        arrowwidth=2,        arrowcolor="#000000",        ax=-10,        ay=-40,        bordercolor="#FFFFFF",        borderwidth=2,        borderpad=4,        bgcolor="#FF5656",        opacity=0.99)

fig.add_annotation(x=57.1, y=53.1,   text="Nigeria", showarrow=True, font=dict(family="Arial", size=16, color="#000000" ),   align="center",      arrowhead=2,        arrowsize=1,        arrowwidth=2,        arrowcolor="#000000",        ax=-10,        ay=-40,        bordercolor="#FFFFFF",        borderwidth=2,        borderpad=4,        bgcolor="#00834e",        opacity=0.99)
 
fig.update_xaxes(title="Índice de libertad económica")
fig.update_yaxes(title="Índice de desarrollo humano")
fig.update_layout(showlegend=False)
fig.update_traces(selector=dict(type='scatter'), showlegend=False)

fig.show()

```

```python
datos = pd.read_excel("corrupcion.xlsx", sheet_name="Merged data")
datos = datos.dropna()


fig = go.Figure()
trendline = px.scatter(datos, x="Transparency", y="Quality of life", trendline="ols")
results = px.get_trendline_results(trendline)
trendline = trendline.data[1]
trendline["marker_color"] = "#FA636E"
fig.add_trace(go.Scatter( x = datos["Transparency"], y = datos["Quality of life"], mode = "markers", text=datos["Nation or Territory"], showlegend=False))
fig.add_trace(go.Scatter(trendline))
fig.update_xaxes(title="Transparency Index")
fig.update_yaxes(title="Quality of life Index")
fig.update_layout(title="Transparency vs. Quality of life")
fig.show()

correlation = np.corrcoef(datos["Transparency"],datos["Quality of life"])
print(correlation[0,1])
results.px_fit_results.iloc[0].summary()
```
